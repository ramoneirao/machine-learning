# **Adequação dos Tipos de Dados**

Este notebook tem como objetivo demonstrar o processo de pré-processamento de dados, com foco na adequação dos tipos de dados. Durante a análise de dados, é comum encontrar inconsistências nos tipos de variáveis (como números armazenados como strings ou valores categóricos representados de forma inadequada). 

O pré-processamento adequado é essencial para garantir que os dados estejam no formato correto para análise e modelagem. Neste notebook, serão abordados os seguintes tópicos:

1. **Identificação de Tipos de Dados**: Verificação dos tipos de dados presentes no dataset.
2. **Conversão de Tipos**: Ajuste dos tipos de dados para os formatos apropriados (int, float, str, etc.).
3. **Tratamento de Inconsistências**: Correção de valores que possam estar fora do padrão esperado.

Este processo é fundamental para garantir a qualidade dos dados e o sucesso das etapas subsequentes de análise e modelagem.
