# **Seleção de Features com Correlação**

A seleção de features é uma etapa essencial no pré-processamento de dados, pois ajuda a identificar e remover variáveis irrelevantes ou redundantes, melhorando a eficiência e a performance dos modelos de machine learning.

Neste notebook, será abordada a **seleção de features baseada em correlação**, com os seguintes tópicos:

1. **Cálculo da Correlação**:
   - Identificação de relações entre variáveis utilizando coeficientes de correlação (como Pearson, Spearman ou Kendall).
   - Análise de correlações positivas e negativas.

2. **Remoção de Features Altamente Correlacionadas**:
   - Identificação de variáveis com alta correlação (multicolinearidade).
   - Eliminação de uma das variáveis correlacionadas para evitar redundância.

3. **Visualização da Correlação**:
   - Uso de heatmaps para visualizar a matriz de correlação.
   - Identificação de padrões e relações entre as variáveis.

A seleção de features baseada em correlação é uma técnica simples e eficaz para reduzir a dimensionalidade do dataset, melhorando a interpretabilidade e a performance dos modelos.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 400)
```

```python
df = pd.read_csv('../base/diabetes.csv')
print(df.corr())
```

```python
# rename das colunas para facilitar a leitura]
df.rename(columns={
    "Pregnancies" : 'preg',
    'Glucose':  'plas',
    'BloodPressure' : 'pres',
    'SkinThickness' : 'skin',
    'Insulin' : 'test',
    'BMI' : 'pedi',
    'Age' : 'age',
    'Outcome' : 'class',
    'DiabetesPedigreeFunction' : 'pedigree'
}, inplace=True)

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), cmap='coolwarm')
plt.show()
```
