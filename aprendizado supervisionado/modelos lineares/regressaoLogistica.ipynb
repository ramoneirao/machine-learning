# **Regressão Logistica**

Estudo de regressão logistica, utilizando o conjunto de dados [Deodorant Instant Liking](https://www.kaggle.com/datasets/ramkumarr02/deodorant-instant-liking-data).


```python
import pandas as pd
from utils import *
from sklearn.model_selection import cross_val_score, StratifiedKFold, GridSearchCV
from sklearn.linear_model import LogisticRegression

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('../../base/DeodorantInstantLiking.csv')
df.head()
```

```python
verifica_faltantes(df)
```

```python
# Excluido as colunas que não são relevantes para a análise
df.drop(columns=['Respondent.ID', 'Product.ID', 'Product', 'q1_1.personal.opinion.of.this.Deodorant', 'q8.20', 'q8.18', 'q8.17', 'q8.8', 'q8.9', 'q8.10', 'q8.2'], inplace=True)

# Completando os valores faltantes com a media das 
df['q8.12'] = df['q8.12'].fillna(df['q8.12'].median())
df['q8.7'] = df['q8.7'].fillna(df['q8.12'].median())
```

```python
verifica_faltantes(df)
```

```python
# Separando 
X = df.drop(columns=['Instant.Liking'])
y = df['Instant.Liking']
```

```python
# Gerando FOLDS
skf = StratifiedKFold(n_splits=5, shuffle=True)

# Criando o modelo
model = LogisticRegression(max_iter=1000)
score = cross_val_score(model, X, y, cv=skf)

print(f'Acurácia: {score.mean():.2f}')
```

```python
model = LogisticRegression(max_iter=1000)

# Definindo os parâmetros para o GridSearch
param_grid = {
    'C': [0.01, 0.1, 1, 2, 3, 5, 10, 20, 50],
    'penalty': ['l1', 'l2'],
    'solver': ['saga'],
}

# Criando o GridSearch
grid = GridSearchCV(estimator = model, param_grid=param_grid, cv=5)
grid.fit(X, y)
```

```python
# imprmindo os melhores parâmetros
print(f'Melhor acuracia: {grid.best_score_:.2f}')
print(f'Melhor parametro C: {grid.best_params_["C"]}')
print(f'Melhor parametro penalty: {grid.best_params_["penalty"]}')
print(f'Melhor parametro solver: {grid.best_params_["solver"]}')
```
