# **Regressão Linear**

Este notebook é dedicado ao estudo e implementação de modelos de regressão linear. Aqui exploraremos conceitos teóricos e práticos, utilizando bibliotecas populares para análise de dados e aprendizado de máquina.

## **Importações**

```python
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
```

## **Geração de Dados Fictícios**

Nesta seção, iremos gerar dados fictícios que serão utilizados para aplicar o modelo de regressão linear. Esses dados serão criados de forma controlada para facilitar a análise e validação do modelo. 

```python
# Gerando dados sintéticos
x, y = make_regression(n_samples=500, n_features=1, noise=30)

# Plotando os dados
plt.scatter(x, y)
plt.show()
```

## **Criação do Modelo**

Nesta seção, iremos criar um modelo de Regressão Linear, após a criação do modelo, realizaremos o treinamento (fit) utilizando os dados fictícios gerados anteriormente. O objetivo é ajustar o modelo para que ele consiga prever os valores de saída (y) com base nos valores de entrada (x).

```python
model = LinearRegression()

# Treinando o modelo
model.fit(x, y)
```

```python
# Coeficientes do modelo
coef_lin = model.intercept_
coef_angular = model.coef_[0]

# Mostrando os coeficientes
print(f'Coeficiente linear: {coef_lin}')
print(f'Coeficiente angular: {coef_angular}')
```

## **Resultados do Modelo**

Nesta seção, iremos exibir os coeficientes do modelo de Regressão Linear treinado. Esses coeficientes incluem o coeficiente linear e o coeficiente angular, que representam a equação da reta ajustada aos dados.

```python
plt.scatter(x, y)
plt.plot(x, coef_angular * x + coef_lin, color='red')
plt.show()
```

## **Aplicando com a Divisão de Treino e Teste**

Nesta seção, iremos dividir os dados em conjuntos de treino e teste para avaliar o desempenho do modelo de Regressão Linear. A divisão permite verificar como o modelo generaliza para dados que não foram utilizados durante o treinamento.

```python
# Dividindo os dados em treino e teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Treinando o modelo com os dados de treino
model.fit(x_train, y_train)
```

```python
# Resultados do modelo
result = model.score(x_test, y_test)
print(f'R²: {result}')
```

```python
# Mostrando os coeficientes
print(f'Coeficiente linear: {model.intercept_}')
print(f'Coeficiente angular: {model.coef_[0]}')
```

```python
# Plotando os dados de teste e a linha de regressão do modelo
plt.scatter(x_test, y_test)
plt.plot(x_test, coef_angular * x_test + coef_lin, color='red')
plt.show()
```
